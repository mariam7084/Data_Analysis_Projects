## Machine Learning Search Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
from pytrends.request import TrendReq
trends=TrendReq()
```

```python
#analyzing the Google search trends on the queries based on "machine learning"

trends.build_payload(kw_list=["Machine Learning"])
df=trends.interest_by_region(resolution='COUNTRY')

df=df.sort_values(by="Machine Learning",ascending=False)
df=df.head(10)

df
```

```python
#visualizing the above data using bar plots

df.reset_index().plot(x='geoName',y='Machine Learning',figsize=(15,10),kind='bar')

plt.show()
```

```python
#analysing trend of searches to see total search queries based
#on "machine learning" increased or decreased over time

#getting data
df=TrendReq(hl='en-US',tz=360)
df.build_payload(kw_list=['Machine Learning'])
df=df.interest_over_time()

#plotting data
fig,ax=plt.subplots(figsize=(15,10))
df['Machine Learning'].plot()

plt.title("Total Google Searches for Machine Learning",fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Total Count")
plt.show()
```
